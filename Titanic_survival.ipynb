```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

train = pd.read_csv("C:/Users/Ank/Desktop/titanic/train.csv")
test = pd.read_csv("C:/Users/Ank/Desktop/titanic/test.csv")
print(train.shape)
print(test.shape)
print(train.head())
print(test.head())
```

```python
print(train.columns)
print(test.columns)
print(train['Sex'].value_counts())
pd.pivot_table(train,index='Sex',values='Survived',aggfunc=np.sum)
```

```python
full_data = [train, test]
full_data
```

```python
print(train.isnull().sum())
print()
print(test.isnull().sum())
```

```python
for data in full_data:
    data['Age']=data['Age'].fillna(data['Age'].mean())
    data['Fare']=data['Fare'].fillna(data['Fare'].median())
    
```

```python
print(train.isnull().sum())
print()
print(test.isnull().sum())
train['Age']
```

```python
for data in full_data:
    data['Sex'] = data['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
    data['Embarked'] = data['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
```

```python
plt.hist(train['Age'])
plt.hist(train['Fare'])
```

```python
for data in full_data:
    data.loc[data['Age']<=15,'Age']=0
    data.loc[(data['Age']>15) & (data['Age']<=25),'Age']=1
    data.loc[(data['Age']>25) & (data['Age']<=35),'Age']=2
    data.loc[(data['Age']>35) & (data['Age']<=45),'Age']=3
    data.loc[(data['Age']>45) & (data['Age']<=55),'Age']=4
    data.loc[data['Age']>55,'Age'] = 5
    data.loc[data['Fare']<=7.91,'Fare']=0
    data.loc[(data['Fare']>7.91) & (data['Fare']<=14.5),'Fare']=1
    data.loc[(data['Fare']>14.5) & (data['Fare']<=31),'Fare']=2
    data.loc[(data['Fare']>31),'Fare']=3
```

```python
train_model = train[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Survived']]
test_model = test[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
train_model.corr()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
m = lr.fit(train_model[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']], train_model['Survived'])
pred = lr.predict(test_model)
pred
```

```python
f = open("C:/Users/Ank/Desktop/titanic/submissionLR.txt","w+")
f.write('PassengerId,Survived\n')
for i,e in enumerate(pred):
    f.write(str(int(892)+i))
    f.write(',')
    f.write(str(e))
    f.write('\n')
f.close()
```

```python
from sklearn.naive_bayes import GaussianNB
lr = GaussianNB()
m = lr.fit(train_model[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']], train_model['Survived'])
pred = lr.predict(test_model)
f = open("C:/Users/Ank/Desktop/titanic/submissionNB_.txt","w+")
f.write('PassengerId,Survived\n')
for i,e in enumerate(pred):
    f.write(str(int(892)+i))
    f.write(',')
    f.write(str(e))
    f.write('\n')
f.close()
```

```python
from sklearn.svm import SVC
lr = SVC()
m = lr.fit(train_model[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']], train_model['Survived'])
pred = lr.predict(test_model)
f = open("C:/Users/Ank/Desktop/titanic/submissionSVC_.txt","w+")
f.write('PassengerId,Survived\n')
for i,e in enumerate(pred):
    f.write(str(int(892)+i))
    f.write(',')
    f.write(str(e))
    f.write('\n')
f.close()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lr = LinearDiscriminantAnalysis()
m = lr.fit(train_model[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']], train_model['Survived'])
pred = lr.predict(test_model)
f = open("C:/Users/Ank/Desktop/titanic/submissionLDA_.txt","w+")
f.write('PassengerId,Survived\n')
for i,e in enumerate(pred):
    f.write(str(int(892)+i))
    f.write(',')
    f.write(str(e))
    f.write('\n')
f.close()
```
